# Modelos ARMA

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (10, 5)
````

O comando `np.random.randn(n)` retorna uma amostra de tamanho `n` normalmente distribuida com média zero e variância 1.

Para amostras aleatórias com distribuição $N(\mu, \sigma^2)$, devemos executar:
```mu + sigma * np.random.randn(...)```

**Série com média constante**

````python
# Vamos criar uma série com 100 valores de uma normal com média 10 e desvio-padrão 0.2
# semente aleatória
np.random.seed(33) 
# vetor com 100 observações
rand = 10 + 0.2 * np.random.randn(100)
````

````python
# calcula uma série com valores médios acumulados do primeiro valor até o dia anterior
mean = [np.mean(rand[:x]) for x in range(len(rand))]

# Plotar gráfico
plt.plot(rand, label='Valor')
plt.plot(mean, label='Média')
plt.title('Série com Média constante')
plt.legend()
plt.show()
````

**Série com média variável**

````python
# Vamos criar uma série com 100 valores de uma normal com média 10 e desvio-padrão 0.2
# semente aleatória
np.random.seed(33) 
# vetor com 100 observações
rand2 = np.cumsum([x for x in range(100) * np.random.randn(100) * 0.2])

#Criando uma média que se expande do primeiro valor e acumulado até cada dia anterior
mean2 = [np.mean(rand2[:x]) for x in range(len(rand2))]

# Plotar gráfico
plt.plot(rand2, label='Valor')
plt.plot(mean2, label='Média')
plt.title('Série com Média variável')
plt.legend()
plt.show()
````

**Série com variância variável**

````python
# Vamos criar uma série com 100 valores de uma normal com média 10 e desvio-padrão 0.1
# semente aleatória
np.random.seed(33) 
# vetor com 100 observações
rand3 = [x for x in range(100) * np.random.randn(100) * 0.1]

#Criando uma média que se expande do primeiro valor e acumulado até cada dia anterior
var = [np.var(rand3[:x]) for x in range(len(rand3))]

# Plotar gráfico
plt.plot(rand3, label='Valor')
plt.plot(var, label='Variância')
plt.title('Série com variância constante')
plt.legend()
plt.show()
````

## Modelos Autoregressivos (AR)

**Modelos AR**

- AR(1): $y_t = \phi_1 y_{t-1} + \epsilon_t$

- AR(2): $y_t = \phi_1 y_{t-1} + \phi_2 y_{t-2} + \epsilon_t$

- AR(p): $y_t = \phi_1 y_{t-1} + \phi_2 y_{t-2} + \cdots + \phi_p y_{t-p} + \epsilon_t$


Podemos reescrever um processo AR(1) como um MA($\infty$):

\begin{equation} 
y_t = c + \phi_1 y_{t-1} + \epsilon_t \\ 
y_t - \phi_1 y_{t-1} =  c + \epsilon_t \\
(1-\phi L) y_t = c + \epsilon_t \\
y_t = \frac{c}{(1-\phi)} + \sum_{j=0}^{\infty} \phi^j \epsilon_{t-j} \\
y_t = \mu + \Phi(L)\epsilon_{t}
\end{equation}

Com isso podemos verificar condições de estabilidade da série calculando seu valor esperado e variância:

\begin{align}
\mathbb{E}[y_t] &= \mu; \\
\mathbb{Var}[y_t] &= \mathbb{E}[y_t - \mu]^2  = \mathbb{E} \left( \sum_{j=0}^{\infty} \phi^j \epsilon_{t-j} \right) =
                    \sum_{j=0}^{\infty} \phi^{2j} \mathbb{E}(\epsilon_{t-j}^2) = \frac{\sigma^2}{1-\phi^2}
\end{align}

Por sua vez, a autocovariância é:
\begin{align*}
\mathbb{E} [(y_{t} - \mu)(y_{t-j} - \mu)] &= 
\mathbb{E} \left[ \left( \sum_{s=0}^{\infty} \phi^{s} \epsilon_{t-s} \right)
  \left( \sum_{s=0}^{\infty} \phi^{s} \epsilon_{t-s-j} \right)  \right] \\
                  &= \sigma^2 \left(\phi^{j} + \phi^{j+2} + \phi^{j+2} + \phi^{j+4} + \cdots \right) \\ 
                  &= \left( \frac{\phi^{j}}{1-\phi^2} \right) \sigma^2
\end{align*}

Note que média e as autocovariância não são funções do tempo.
A autocorrelação de ordem $j$ é dada por:

$$\rho_j = \frac{ \left(  \frac{\phi^{j}}{1-\phi^2} \right) \sigma^2 }{ \frac{\sigma^2}{1-\phi^2} } = \phi^{j}, \space j =1, 2, \cdots $$

Para simular processos ARMA vamos usar o módulo `arima_process` no `statsmodels`.

````python
# Importar arma_process
from statsmodels.tsa.arima_process import ArmaProcess

# Importar plot_acf
from statsmodels.graphics.tsaplots import plot_acf
````

Vamos começar simulando um processo AR(1), verificando o gráfico da série e o seu correlograma.

### AR(1)

````python
#  AR(1) com parâmetro = + 0.9
ar1 = np.array([1, -0.9]) # note o sinal inverso 
# Componente simples 
ma1 = np.array([1])

ar1_process = ArmaProcess(ar1, ma1)
sim_ar1 = ar1_process.generate_sample(nsample=100)

# Plotar série
plt.plot(sim_ar1)
plt.title('AR(1), phi = 0.9')
# Plotar correlograma do processo AR(1), parâmetro = +0.9
plot_acf(sim_ar1, alpha=1, lags=20)
plt.title('AR(1), phi = 0.9')
plt.show()
````

Vamos ver o comportamento do processo AR(1).

````python
#  AR(1) com parâmetro = - 0.9
ar1b = np.array([1, 0.9]) # note o sinal inverso 
# Componente simples 
ma1 = np.array([1])

ar1_process2 = ArmaProcess(ar1b, ma1)
sim2_ar1 = ar1_process2.generate_sample(nsample=100)

# Plotar série
plt.plot(sim2_ar1)
plt.title('AR(1), phi = - 0.9')
plt.show()
# Plotar correlograma do processo AR(1), parâmetro = - 0.9
plot_acf(sim2_ar1, alpha=1, lags=20)
plt.title('AR(1), phi = - 0.9')
plt.show()
````

### Modelos de médias móveis (MA)

**Modelos MA**

- MA(1): $y_t = \theta_1 \epsilon_{t-1} + \epsilon_t$

- MA(2):: $y_t = \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \epsilon_t$

- MA(q): $y_t = \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \cdots + \theta_q \epsilon_{t-q} + \epsilon_t$


$$


````python
#  MA(1) com parâmetro = - 0.9
ar1 = np.array([1])
ma1 = np.array([1, -0.9])
ma1_process = ArmaProcess(ar1, ma1)
sim1_ma1 = ma1_process.generate_sample(nsample=200)
# plotar
plt.subplot(2, 1, 1)
plt.plot(sim1_ma1);

#  MA(1) com parâmetro = + 0.9
ma1b = np.array([1, 0.9])
ma1_process2 = ArmaProcess(ar1, ma1b)
sim2_ma1 = ma1_process2.generate_sample(nsample=200)
# plotar
plt.subplot(2, 1, 2)
plt.plot(sim2_ma1);
````

Diferente do AR(1), um modelo MA(1) não tem autocorrelação além do lag 1. Por sua vez, um modelo MA(2) não tem autocorrelação além do lag 2, e assim por diante. A autocorrelação lag-1 para um modelo MA(1) não é $\theta$, mas sim $$\frac{\theta}{1+\theta^2}$$

````python
from statsmodels.graphics.tsaplots import plot_acf

# Plotar correlograma do processo MA(1), parâmetro = - 0.9
plot_acf(sim1_ma1, lags=20);

# Plotar correlograma do processo MA(1), parâmetro = + 0.9
plot_acf(sim2_ma1, lags=20);
````
